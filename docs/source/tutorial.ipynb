# Tutorial

To start using the `memprofiler` extension in a Jupyter notebook, simply run:

```python
%load_ext memprofiler
```

In this tutorial we will go through all the options of the memprofiler extension
while working with numpy arrays. So let's get started!

```python
import numpy as np
```

```python
shape = (3000, 3000)
a = np.linspace(0, 1, np.prod(shape)).reshape(shape)
```

## Start sampling memory

```python
%%mprof_run expr_sin

_ = np.sin(a)
```

```python
%%mprof_run -i 0.002 expr_cos

_ = np.cos(a)
```

```python
%%mprof_run --plot expr_tan

_ = np.sin(a) / np.cos(a)
```

## Plot multiple memory profiles at once

```python
%mprof_plot expr_sin expr_cos expr_tan
```

```python
%mprof_plot expr.*
```

```python
%mprof_plot --title "Numpy Trigonometric Functions" expr.*
```
